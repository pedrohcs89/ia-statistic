```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import locale

locale.setlocale(locale.LC_TIME, 'pt_BR.utf8')
```

```python
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

```python
df_faturamento = pd.DataFrame.from_dict(dict_faturamento)
df_faturamento.info()
```

```python
media_vendas = round(df_faturamento["valor"].mean(), 2)
print(f"Média das vendas = {media_vendas}")
```

```python
df_faturamento["data_ref"] = pd.to_datetime(df_faturamento["data_ref"])
df_faturamento.info()
```

```python
data_mes = df_faturamento["data_ref"].dt.strftime("%b")
valor = df_faturamento["valor"]
```

```python
plt.bar(data_mes, valor)

ax = plt.gca()
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"{x:,.0f}".replace(",", ".")))

plt.xlabel("Mês")
plt.ylabel("Valor")
plt.title("Faturamento por Mês")
```
